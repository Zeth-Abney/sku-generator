# Spreadsheet clean up for Meta catalog upload
remove html tags from descriptions

```python
file_path = '/Users/zethu/Desktop/Legendary Concepts/catalog_products.numbers'
```

```python
import pandas as pd
from numbers_parser import Document
import re
import html

# Load the Numbers file
doc = Document(file_path)
sheets = doc.sheets

# Find the Meta table
table = None
for sheet in sheets:
    for t in sheet.tables:
        if t.name == 'meta_catalog_products':
            table = t
            break
    if table:
        break

if table is None:
    raise ValueError("Could not find 'Meta' table in the Numbers file")

# Convert to DataFrame
data = table.rows(values_only=True)
df = pd.DataFrame(data[1:], columns=data[0])

# Function to remove HTML tags and entities
def remove_html_tags(text):
    if pd.isna(text) or not isinstance(text, str):
        return text
    # Remove HTML tags
    clean = re.sub('<.*?>', '', text)
    # Decode HTML entities like &nbsp; &amp; etc.
    clean = html.unescape(clean)
    # Remove any remaining HTML entities (sometimes double-encoded)
    clean = re.sub('&[a-zA-Z]+;', ' ', clean)
    # Clean up extra whitespace
    clean = ' '.join(clean.split())
    return clean

# Clean the description column
df['description'] = df['description'].apply(remove_html_tags)

# Display the cleaned data
df
```

```python
# Save the cleaned data back to the Numbers file
# Update the table with cleaned data
for i, row in df.iterrows():
    for j, col in enumerate(df.columns):
        value = row[col]
        # Convert None/NaN to empty string for numbers_parser
        if pd.isna(value) or value is None:
            value = ""
        table.write(i + 1, j, value)

# Save the document
doc.save(file_path)
print(f"✓ HTML tags removed from description column and saved to {file_path}")
```
